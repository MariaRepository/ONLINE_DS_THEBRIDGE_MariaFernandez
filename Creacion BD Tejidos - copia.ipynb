```python
# Creamos la tabla Pedido
query = '''
CREATE TABLE Pedido (
    ID_Pedido INTEGER PRIMARY KEY,
    Fecha_pedido DATE,
    Fecha_entrega DATE,
    ID_Proveedor INTEGER,
    FOREIGN KEY (ID_Proveedor) REFERENCES Proveedor(ID_Proveedor)
    )
'''
cursor.execute(query)
```

```python
# # Generar datos para la tabla PEDIDO

import datetime

# Generar 20 valores 
valores_pedidos = []
for i in range(20):
    Id_pedido = i + 1 
    fecha_pedido = datetime.datetime.now().strftime('%Y-%m-%d')
    fecha_entrega = (datetime.datetime.now() + datetime.timedelta(days=random.randint(1, 30))).strftime('%Y-%m-%d')
    id_proveedor = random.randint(1, 20)
    valores_pedidos.append((Id_pedido, fecha_pedido, fecha_entrega, id_proveedor))

# Imprimir los valores generados
for idx, valor in enumerate(valores_pedidos, start=1):
    print(f"{idx}. {valor[0]} Fecha Pedido: {valor[1]}, Fecha Entrega: {valor[2]}, ID_Proveedor: {valor[3]}")
```

```python
# Insertar datos en la tabla Pedido
insert_query_pedidos = "INSERT INTO Pedido (ID_Pedido, Fecha_pedido, Fecha_entrega, ID_Proveedor) VALUES (?, ?, ?, ?)"

# Ejecutar la sentencia SQL para insertar los datos en la tabla
cursor.executemany(insert_query_pedidos, valores_pedidos)

# Confirmar la transacción
conexion.commit()
```

```python
# Creamos la tabla Catalogo
query = '''
CREATE TABLE Catalogo (
    ID_Producto INTEGER,
    ID_Proveedor INTEGER,
    Precio REAL,
    FOREIGN KEY (ID_Producto) REFERENCES Producto(ID_Producto)
    FOREIGN KEY (ID_Proveedor) REFERENCES Pedido(ID_Proveedor)
    CONSTRAINT ID_Catalogo PRIMARY KEY (ID_Producto, ID_Proveedor)   
    )
'''
cursor.execute(query)
```

```python
# Consulta de los tejidos de una categoria con su proveedor y precio
query1 = '''
SELECT a.Nombre As CATEGORIA, b.Nombre AS TEJIDO, b.Color As COLOR, c.Precio AS PRECIO, d.Nombre AS PROVEEDOR
FROM Categoria AS a
INNER JOIN Producto AS b ON (a.ID_Categoria = b.ID_Categoria)
INNER JOIN Catalogo AS c ON (b.ID_Producto = c.ID_Producto)
INNER JOIN Proveedor AS d ON (c.ID_Proveedor = d.ID_Proveedor)
WHERE a.Nombre = "Fibra sintética"
ORDER BY b.Nombre

'''
df_query1 = pd.read_sql(query1, conexion)
df_query1  
```

```python
# Consulta de las cantidades entregadas de un tejido su proveedor y fecha de entrega

query2 = '''
SELECT a.Nombre AS TEJIDO, a.Color As COLOR, c.ID_Pedido AS NUM_PEDIDO, c.Fecha_entrega AS FECHA_ENTREGA, b.Cantidad AS CANTIDAD, d.Nombre AS PROVEEDOR
FROM Producto AS a
INNER JOIN LineaCesta AS b ON (a.ID_Producto = b.ID_Producto)
INNER JOIN Pedido AS c ON (b.ID_Pedido = c.ID_Pedido)
INNER JOIN Proveedor AS d ON (c.ID_Pedido = d.ID_Proveedor)
WHERE a.Nombre = "Mohair Exclusivo"
ORDER BY c.Fecha_entrega

'''
df_query2 = pd.read_sql(query2, conexion)
df_query2  
```
