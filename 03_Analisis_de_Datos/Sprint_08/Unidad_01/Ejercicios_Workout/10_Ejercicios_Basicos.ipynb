### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
fig = plt.figure()
plt.xlim(0.00,10.0)
plt.ylim(-1.00,1.00)

plt.subplot(1, 2, 1) # filas, columnas, indice
plt.plot(x, np.sin(x),"blue",linestyle='solid')
plt.title("Seno");



plt.subplot(1, 2, 2) # filas, columnas, indice
plt.plot(x, np.cos(x))
plt.title("Coseno");
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python

fig = plt.figure()

plt.ylim(0,10000);

plt.subplot(2, 1, 1) # filas, columnas, indice
plt.plot(x, np.sinh(x),'r')
plt.title("Seno hiperbólico")
plt.xlabel("x")
plt.ylabel("sinh(x)");

plt.subplot(2, 1, 2) # filas, columnas, indice
plt.plot(x, np.cosh(x),'b')
plt.title("Coseno hiperbólico")
plt.xlabel("x")
plt.ylabel("sinh(x)");
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python

fig3 = plt.figure()
fig.subplots_adjust(hspace=10, wspace= 0.5)

plt.subplot(2, 2, 1) # filas, columnas, indice
plt.plot(x, np.sin(x));


plt.subplot(2, 2, 2) # filas, columnas, indice
plt.plot(x, np.cos(x));


plt.subplot(2, 2, 3) # filas, columnas, indice
plt.plot(x, np.tan(x));


plt.subplot(2, 2, 4) # filas, columnas, indice
plt.plot(x, np.sqrt(x));
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando sólo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros gráficos (fila 1). Luego usa el estilo con métodos (objetos `axes`) para añadir el tercer gráfico.

```python
fig4 = plt.figure()
fig.subplots_adjust(hspace=1.0, wspace= 0.5)



plt.subplot(2, 1, 1) # filas, columnas, indice
plt.plot(x, np.cos(x),"blue",linestyle='solid',label='coseno')
plt.title("Coseno")
plt.xlim(0.00,10.0)
plt.ylim(-1.00,1.00);

plt.subplot(2, 1, 2) # filas, columnas, indice
plt.plot(x, np.sin(x),"blue",linestyle='solid')
plt.title("Seno")
plt.xlim(0.00,10.0)
plt.ylim(-1.00,1.00);

plt.legend()

ax3 = fig.add_subplot(2, 1, 1, sharex=ax3)  # Agregar un tercer subplot en la misma posición que el segundo

ax3.plot(x, np.sqrt(x), "g", linestyle='solid')
ax3.set_xlim(0.00, 10.0) 
ax3.set_ylim(-1, 1) ;
```
